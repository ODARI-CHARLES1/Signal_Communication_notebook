```python
import sympy as sp
import scienceplots
import matplotlib.pyplot as plt

plt.style.use(["science","no-latex"])
```

```python
from sympy.abc import t,n
```

```python
f=sp.Heaviside(4*t+1.5)-sp.Heaviside(4*t-1.5)
```

```python
sp.plot(f,xlim=[-0.5,0.5])
plt.show()
```

```python
g=f.subs(t,t-4)
sp.plot(g,xlim=[0,8])
```

```python
sp.plot(g.subs(t,4*t),xlim=[0,3])
plt.show()
```

```python
sp.plot(sp.exp(-4*t)*sp.Heaviside(t),xlim=[-1,1],ylim=[0,2])
```

```python
g=sp.exp(-t)*sp.Heaviside(t)
```

```python
sp.plot(g,xlim=[-5,5])
```

```python
odd=(g-g.subs(t,-t))/2
odd
```

```python
sp.plot(odd,xlim=[-5,5])
```

```python
even=(g+g.subs(t,-t))/2
sp.simplify(even)
```

```python
sp.plot(even,xlim=[-5,5])
```

```python
sp.plot(even+odd,xlim=[-5,5])
plt.show()
```
