```python
import sympy as sp
```

```python
from sympy.abc import x
```

```python
f=sp.sin(x)
```

```python
sol=sp.fourier_series(f,(x,-sp.pi,sp.pi))
```

```python
sol.truncate(1)
```

```python
f=x
```

```python
sp.plot(f)
```

```python
sol=sp.fourier_series(f,(x,-sp.pi,sp.pi))
p=sol.truncate(10)
p
```

```python
sp.plot(p,xlim=[0,sp.pi])
```

```python
sp.integrate(x,(x,-sp.pi,sp.pi))
```
