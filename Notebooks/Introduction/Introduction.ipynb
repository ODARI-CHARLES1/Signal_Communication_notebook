# Imports

```python
import numpy as np
from scipy import signal 
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.express as px 
import os 
import pathlib
from warnings import filterwarnings
import scienceplots
import sympy as sp

filterwarnings("ignore")
plt.style.use(['science', 'ieee']) # Requires scienceplots package

# Disable external LaTeX dependency
plt.rcParams['text.usetex'] = False
plt.rcParams['font.family'] = 'DejaVu Sans' 

# Figure settings
plt.rcParams['figure.figsize'] = (10,4)
plt.rcParams['figure.dpi'] = 120
```

# Plot a sinusoidal wave

$y(t)=Asin(2πft)$

 where
 $A=0.1 f=1$

```python
t=np.linspace(0,4*np.pi,1000)
f=1 # 1Hz
A=0.1

y1=A*np.sin(2*np.pi*f*t)
y2=A*np.cos(2*np.pi*f*t)


plt.plot(t, y1, color="#1f77b4", linewidth=2, label="0.1 sin(2πft)")
plt.plot(t,y2,color='red',linewidth=2,label="0.1 cos(2πft")

# Titles and labels (clean + readable)
plt.title("Sinusoidal Signal (Time Domain)", fontsize=14, fontweight="bold")
plt.xlabel("Time (s)", fontsize=12)
plt.ylabel("Amplitude (V)", fontsize=12)

# Grid (engineering style: subtle but informative)
plt.grid(True, linestyle="--", linewidth=0.6, alpha=0.7)

# Axes styling
plt.axhline(0, color="black", linewidth=0.8)
plt.xlim(0, 4)

# Legend
plt.legend(loc="upper right")

# Remove top/right borders (clean paper style)
ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.show()
```

## Continouous and Discrete Time Signals

```python
#defining the step plot values
n=np.linspace(0,4*np.pi,100)
y_stem=A*np.sin(2*np.pi*f*n)


#Plot of continouous signal
fig,ax=plt.subplots(ncols=1,nrows=2,figsize=(10,8))
ax[0].plot(t,y1,"r")
ax[0].grid(True,linestyle="--",alpha=0.4)
ax[0].set_title("Continuous Time Signal")

#plot of discrete signnal
ax[1].stem(n,y_stem,"b")
ax[1].set_title("Discrete Time Signals")

#setting the grid, x and y label

for index,ax in enumerate(ax):
    ax.set_xlabel("time(s)")
    ax.set_ylabel("Amplitude")
    ax.grid(True,linestyle="--",alpha=0.3)
    
```

## Square wave plot

```python

#parameters
pscaling=2+

nscaling=0.5


t=np.linspace(0,1,500,endpoint=False)
y_square=signal.square(2*np.pi*3*t)



fig,ax=plt.subplots(nrows=3,ncols=1,figsize=(10,8))

#origin signal
ax[0].plot(t,y_square,"orange")
ax[0].set_title("Unscaled time")

#scaling by a factor greater than 1
ax[1].plot(t,signal.square(2*np.pi*3*t*pscaling),"red")
ax[1].set_title("Time scaled by a factor greater than 1")

#scaling by factor less than 1
ax[2].plot(t,signal.square(2*np.pi*3*t*nscaling),"blue")
ax[2].set_title("Time scaled by a factor less than 1")
           
for index,ax in enumerate(ax):
    ax.set_xlabel("time(s)")
    ax.set_ylabel("Amplitude")
    ax.grid(True,linestyle="--",alpha=0.3)

plt.tight_layout()
plt.show()
```
