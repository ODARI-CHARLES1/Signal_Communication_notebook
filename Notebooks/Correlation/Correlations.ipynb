# Inner Product and Norm of Signals

## Inner Product

$
\begin{equation}
    \langle g(t), x(t) \rangle = \int_{\Theta} g(t)x^*(t) \, dt
\end{equation}
$

## Norm of a signal

$ ||g(t)||=[ \int_{t:t\in \Theta}^{} g(t)x^*(t)dt] ^{\frac{1}{2}}$

```python
#plot of the signals between 0 and 2*pi
t_values=np.linspace(0,2*np.pi,1000)
fig,ax=plt.subplots(nrows=2,ncols=1,figsize=(10,8))
ax[0].plot(t_values,np.sin(t_values),color="red")
ax[0].set_title("Sin(wt) Plot")
ax[1].plot(t_values,np.cos(t_values),color="green")
ax[1].set_title("Cos(wt) Plot")

for x in ax:
    x.set_xlabel("Time")
    x.set_ylabel("y(t)")
    x.grid(True,alpha=0.5,linestyle='--',linewidth=1)
```
