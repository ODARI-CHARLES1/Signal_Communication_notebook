```python
import numpy as np
import scipy as sp
import seaborn as sns
import matplotlib.pyplot as plt
import scienceplots 

plt.style.use(['science','no-latex'])
```

Compute the fourier series of the following signal x(t):

$x(t)=sin(2\pi*50t)+sin(2\pi*60t)$

```python
t=np.linspace(0,0.1*np.pi,1000)
x=np.sin(2*np.pi*50*t)+np.sin(2*np.pi*200*t)
fig,ax=plt.subplots(figsize=(20,5))
ax.plot(t,x,color="green")
ax.set_title("Plot of x(t)")
ax.set_xlabel("time (s)")
ax.set_ylabel("x(t)")
ax.grid(True,linestyle="--",alpha=0.2)
plt.show()
```

```python

Fs = 1000      # Sampling frequency
T = 1          # Duration (seconds)

t = np.linspace(0, T, Fs, endpoint=False)

x = np.sin(2*np.pi*50*t) + np.sin(2*np.pi*200*t)

X = np.fft.fft(x)

freqs = np.fft.fftfreq(len(x), 1/Fs)

mag = np.abs(X)/len(x)

mask = freqs >= 0
fig,ax=plt.subplots(figsize=(20,5))
ax.plot(freqs[mask], mag[mask])
ax.set_xlabel("Frequency (Hz)")
ax.set_ylabel("Magnitude")
ax.set_title("FFT Spectrum")
ax.grid(True)
plt.show()
```
