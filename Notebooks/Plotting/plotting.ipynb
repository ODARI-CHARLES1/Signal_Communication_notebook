```python
import sympy as sp
```

```python
from sympy.abc import t
```

```python
#defining a unit step function

y1=sp.Piecewise((0,t<0),(1,t>=0))
y2=sp.Piecewise((0,t<0),(t,t>=0))
```

```python
y1
```

```python
y2
```

```python
sp.simplify(y1-y2)
```

```python
y3=sp.Piecewise((0,t<1),(1,t>=1))
```

```python
sp.simplify(y1-y3)
```

```python
y4=sp.Piecewise((0,t<2),(t,t>=2))
```
