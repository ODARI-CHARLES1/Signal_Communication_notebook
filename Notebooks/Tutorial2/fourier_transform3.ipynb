```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import scipy as scp
import scienceplots

plt.style.use(["science","no-latex"])
```

```python
from sympy.abc import t,a,tau
```

```python
def comput_energy(func,variable=t):
    E=sp.integrate(func,(variable,-sp.oo,sp.oo))
    return E
```

```python
y=sp.Heaviside(t+1)-sp.Heaviside(t-1);
y
```

```python
sp.plot(y,xlim=[-2,2])
plt.show()
```

```python
#Determining its energy
Ex=sp.integrate(y,(t,-sp.oo,sp.oo))
Ex
```

```python
x=y.subs(t,2*(t-1))
```

```python
sp.plot(x,xlim=[-2,2])
plt.show()
```

```python
Ex=comput_energy(x)
Ex
```

```python
import numpy as np
import matplotlib.pyplot as plt

tau = np.array([-2, -0.5, 0, 0.5, 1.5, 3])
psi = np.array([0, 0.5, 1, 1, 0.5, 0])

plt.plot(tau, psi, marker='o')
plt.grid(True)
plt.xlabel(r'$\tau$')
plt.ylabel(r'$\psi_{xy}(\tau)$')
plt.title('Cross-Correlation Function')
plt.show()
```
